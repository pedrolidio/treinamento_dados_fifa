## Introdução
Este notebook visa praticar as principais operações  para a manipulação e visualização de dados utilizando python.
As operações utilizadas serão responsáveis por projetar, filtrar, agrupar, pivotear, concatenar e/ou mesclar os dados de uma ou mais tabelas. Abaixo temos os imports necessários para as bibliotecas utilizadas e a importação dos dados de uma das tabelas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

data = pd.read_csv(f'data/players_22.csv', low_memory=False)
```

## Principais operações
### Visualização
Para visualizar os dados de uma tabela, podemos utilizar as funções head ou tail, que mostram respectivamente as n primeiras ou n ultimas linhas de uma tabela.

```python
data.head()
```

```python
data.tail()
```

### Projeção
A operação de projeção é a responsável por selecionar os dados de uma tabela pelas colunas de interesse. Como exemplo, podemos selecionar apenas as colunas que contém o id, nome (curto), idade, posição e valor de um jogador.

```python
data[['sofifa_id', 'short_name', 'age', 'club_position', 'value_eur']]
```

### Filtragem
De forma semelhate à operação de projeção, a operação de filtragem é responsável por selecionar os dados de uma tabela pelas linhas de interesse. Para este fim, podem ser utilizadas as funções query e loc. Como exemplo, podemos selecionar as linhas contendo os jogadores com 30 anos ou mais do Barcelona e do Real Madrid.

```python
data.query("(club_name == 'FC Barcelona' or club_name == 'Real Madrid CF') and age >= 30")[['short_name', 'overall', 'age', 'club_position', 'club_name']]
```

```python
data.loc[((data['club_name'] == 'FC Barcelona') | (data['club_name'] == 'Real Madrid CF')) & (data['age'] >= 30), ['short_name', 'overall', 'age', 'club_position', 'club_name']]
```

### Agrupamento
O agrupamento gera uma nova tabela, cujos dados são relativos a um agrupamento de dados da tabela original. Como exemplo, podemos criar uma tabela contento o número de jogadores no clube, idade média, valor total de jogadores, valor do jogador mais caro e overall médio e máximo dos jogadores dos clubes.

```python
data.groupby(['club_name']).aggregate({
    'sofifa_id': [('#players','count')],
    'age': 'mean',
    'value_eur': [('sum (million)', lambda x: sum(x) / 1e+6), ('max (million)', lambda x: max(x) / 1e+6)],
    'overall': [('mean'),('max')]
}).head()
```

### Pivoteamento
A tabela resultante de um pivoteamento contém as linhas baseadas em um agrupamento e as colunas baseadas nos valores presentes em outra coluna da tabela original. Um exemplo desta operação é a criação de uma tabela contendo o valor médio dos jogadores de cada idade e overall, para os dados do Fifa 20.

```python
data_20 = pd.read_csv(f'data/players_20.csv')

value_per_league_age = data_20.pivot_table(
    values = 'value_eur', 
    index = 'age', 
    columns = 'overall', 
    aggfunc = lambda x: np.mean(x) / 1000
)

value_per_league_age.head(10)
```

Em uma tabela de pivoteamento, muitas vezes é interessante criarmos um mapa de calor (heatmap). Para isto, podemos usar matplotlib ou seaborn para plotagem dos dados.

```python
fig, axes = plt.subplots(1, 2, figsize = (14,6))

sns.heatmap(value_per_league_age, ax = axes[0], cmap = 'coolwarm')

axes[1].set_xlabel('overall')
axes[1].set_ylabel('age')

axes[1].set_xticks(range(value_per_league_age.shape[1])[::5], value_per_league_age.columns[::5])
axes[1].set_yticks(range(value_per_league_age.shape[0])[::5], value_per_league_age.index[::5])

im1 = axes[1].contourf(value_per_league_age, levels = 100, cmap = 'coolwarm')
fig.colorbar(im1)
```

### Despivoteamento
De uma tabela pivoteada, é gerada uma tabela com uma coluna referente às diferentes colunas da tabela pivoteada e outra coluna referente aos valores. Para exemplificar esta operação criaremos uma tabela contendo como colunas overall e valor médio de jogadores para cada idade, baseada na tabela pivoteada anteoriomente.

```python
value_per_league_age.reset_index().melt(id_vars='age', var_name='overall', value_name='mean_value')
```

### Concatenação (union all)
A concatenação consiste na junção de duas tabelas, de modo que as linhas da segunda tabela serão adicionadas ao fim da primeira. Na concatenação, caso existam linhas iguais em ambas as tabelas, não haverá remoção das duplicatas. Para exemplificar, faremos a concatenação da lista de nomes dos jogadores do Paris Saint-Germain com 30 anos ou mais presentes no Fifa 21 e Fifa 22.

```python
data_fifa_21 = pd.read_csv(f'data/players_21.csv')

psg_21 = data_fifa_21.query("club_name == 'Paris Saint-Germain' and age >= 30")[['short_name', 'club_name']]

psg_21.head()
```

```python
psg_22 = data.query("club_name == 'Paris Saint-Germain'and age >= 30")[['short_name', 'club_name']]

psg_22.head(10)
```

Podemos perceber que alguns jogadores aparecem em ambas as tabelas. A tabela resultante da concatenação deverá ter os nomes que aparecem em ambas duplicados.

```python
pd.concat([psg_21,psg_22])
```

### Union
Caso seja de interesse juntar os dados de duas tabelas mas removendo as duplicatas, pode ser utilizada a função drop_duplicates do pandas.

```python
pd.concat([psg_21,psg_22]).drop_duplicates()
```

### Mescla (merge, join)
A mescla realiza o produto cartesiano entre duas tabelas, de forma que a tabela resultante possui dados relacionados por uma chave presente em ambas as tabelas. Um exemplo disto é formar uma tabela contendo o id do jogador, nome do jogador, id do clube, nome do clube e liga do clube, a partir da mescla entre uma tabela contendo os dados dos jogadores e outra contendo os dados dos clubes.

```python
players_22 = data[['sofifa_id', 'short_name', 'club_team_id']]

players_22.head(10)
```

```python
clubs_22 = data[['club_team_id', 'club_name', 'league_name']].drop_duplicates()

clubs_22.head(10)
```

```python
pd.merge(players_22, clubs_22, how='outer',on='club_team_id')
```

A título de curiosidade, a função merge é capaz de realizar diferentes tipos de join, apenas especificando o argumento "how". Este argumento indica o método de merge, sendo capaz de realizar left join, right join, full outer join, inner join e cross join.
